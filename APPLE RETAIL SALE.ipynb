```python
import pandas as pd

# Read all CSV files
category = pd.read_csv('C:/Users/USER/Downloads/archive (1)/category.csv')
products = pd.read_csv('C:/Users/USER/Downloads/archive (1)/products.csv')
sales = pd.read_csv('C:/Users/USER/Downloads/archive (1)/sales.csv')
stores = pd.read_csv('C:/Users/USER/Downloads/archive (1)/stores.csv')

# Check column names first
print("Sales columns:", sales.columns.tolist())
print("Stores columns:", stores.columns.tolist())
print("Products columns:", products.columns.tolist())
print("Category columns:", category.columns.tolist())

# Perform the joins with EXPLICIT join conditions (like your SQL)
we_knows = sales.merge(stores, 
                       left_on='store_id', 
                       right_on='Store_ID', 
                       how='inner') \
                .merge(products, 
                       left_on='product_id', 
                       right_on='Product_ID', 
                       how='inner') \
                .merge(category, 
                       left_on='Category_ID',  # from products table
                       right_on='category_id', 
                       how='inner')

# Select and rename columns to match your SQL output
we_knows = we_knows[[
    'sale_id',
    'sale_date.1',
    'quantity',
    'Store_Name',
    'City',
    'Country',
    'Product_Name',
    'Price',
    'category_name'
]].rename(columns={'sale_date.1': 'sale_date'})

print(f"\nTotal rows: {len(we_knows)}")
print(we_knows.head())

# Save to CSV
we_knows.to_csv('we_knows_full.csv', index=False)
```
